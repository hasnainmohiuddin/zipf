The module `wordcount.py` contains several useful functions for this purpose, so we import them here, along with the numpy and matplotlib libraries that we will use for processing and plotting the collected data:

```python
import wordcount as wc
import numpy as np
import matplotlib.pyplot as plt
```

Next we specify the input file (classical novel from Gutenberg.org in .txt format) that we want to get the word counts for. Use any other such file here if you want to run the analysis for a different novel. Other options in the data/ directory include: frankenstein.txt, jane_eyre.txt, moby_dick.txt, sense_and_sensibility.txt, sherlock_holmes.txt, and time_machine.txt.

```python
inputfile = "../data/dracula.txt"
```

We also specify the name for the outputfile that will store the word counts for the novel:

```python
outputfile = "../data/dracula_counts.txt"
```

Now we can call the word_count function from the wordcound package to get the wordcounts (words, counts and percentages) for our chosen novel: 

```python
wc.word_count(inputfile, outputfile)
```

Having the word counts available, we can now load them again and create a plot that visualiztes the data. (Feel free to play with the parameters to adjust the visualization to your needs.)

```python
# load word counts
counts = wc.load_word_counts(outputfile)

# limit to first 25 words
limited_counts = counts[0:25]

# calculate plot values
word_data = [word for (word, _, _) in limited_counts]
count_data = [count for (_, count, _) in limited_counts]
position = np.arange(len(word_data))
width = 1.0

# create the plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title("Word Counts")
ax.set_xticks(position + (width / 2))
ax.set_xticklabels(word_data)
ax.bar(position, count_data, width, color="b")

# show plot
plt.show()
```
